```python
# importing required packages for the collection of review's
from urllib.request import urlopen as uRequest
from urllib.error import HTTPError
from urllib.error import URLError
from bs4 import BeautifulSoup as soup
import pandas as pd
import re

# read flipcark page of iphone starting page of review system
flipkart_review_page = "https://www.flipkart.com/apple-iphone-x-space-gray-64-gb/product-reviews/itmexrgv6hctyrav?pid=MOBEXRGVCMGVCGGQ&lid=LSTMOBEXRGVCMGVCGGQLHGFDF&marketplace=FLIPKART"   
flipkartLink = "https://www.flipkart.com"
listOfAllDict = []
def getPageReviews(requestLink):
        try:
            # connect and request the page.
            openFlipKartPage = uRequest(requestLink)
             # read and close the request
            readFlipKartpage = openFlipKartPage.read()
            openFlipKartPage.close()
        except HTTPError as e:
            print("Error while reading the page: "+ str(e))
            return None
        except URLError as e:
            print("Error while reading the page: "+ str(e))
            return None
        else:
            pass
        # parse the html page by using soup
        flipkartSoupObj = soup(readFlipKartpage, "html.parser")
        for eacRead in flipkartSoupObj.findAll("div",{"class": "col _390CkK _1gY8H-"}):
            DictFrame = {}
            try:
                productMainReview = eacRead.div.find("p", {"class": "_2xg6Ul"}).text
                DictFrame["productMainReview"] = productMainReview
            except AttributeError as e:
                DictFrame["productMainReview"] = None
                print("productMainReview: "+str(e))
            try:
                productMainRating = eacRead.div.find("div", {"class": "hGSR34 E_uFuv"}).text
                DictFrame["productMainRating"] = productMainRating
            except AttributeError as e:
                DictFrame["productMainRating"] = None
                print("productMainRating: "+ str(e))
            
            try:
                reviewComments = eacRead.find("div",{"class": "qwjRop"}).div.div.text
                DictFrame["reviewComments"] = reviewComments
            except AttributeError as e:
                DictFrame["reviewComments"] = None
                print("reviewComments: "+ str(e))
                
            try:
                customerInfo = eacRead.find("div",{"class":"row _2pclJg"}).p.text
                DictFrame["customerInfo"] = customerInfo
            except AttributeError as e:
                DictFrame["customerInfo"] = None
                print("customerInfo"+ str(e))
                
            try:
                certificationInfo = eacRead.find("p",{"class":"_19inI8"}).text
                DictFrame["certificationInfo"] = certificationInfo
            except AttributeError as e:
                DictFrame["customerInfo"] = None
                print("customerInfo"+ str(e))
                
            try:
                dateInfo = eacRead.findAll("p",{"class": "_3LYOAd"})[1].text
                DictFrame['dateInfo'] = dateInfo
            except AttributeError as e:
                DictFrame['dateInfo'] = None
                print("dateInfo: "+str(e))
            
            try:
                likesInfo, disLinkInfo = tuple(eacRead.findAll("span",{"class": "_1_BQL8"}))
                likesInfo = likesInfo.text
                DictFrame["likesInfo"] = likesInfo
                disLikesInfo = disLinkInfo.text
                DictFrame["disLikesInfo"] = disLikesInfo
            except AttributeError as e:
                DictFrame['likesInfo'] = None
                DictFrame['disLikesInfo'] = None
                print("likes and dislikes: "+ str(e))
            listOfAllDict.append(DictFrame)
        tempFrame = pd.DataFrame(listOfAllDict)
        try:
            if flipkartSoupObj:
                foundList = flipkartSoupObj.find("div", {"class": "_2zg3yZ _3KSYCY"}).findAll("a",{"class": "_3fVaIS"})
                if len(foundList) > 1:
                    nextPageLink = foundList[len(foundList)-1]['href']
                else:
                    nextPageLink = foundList[0]['href']
                    
            if nextPageLink and foundList[len(foundList)-1].span.text == 'Next':
                nextLink = str(flipkartLink)+str(nextPageLink)
                print(nextLink)
                finalFrame = getPageReviews(nextLink)
                return finalFrame
            else:
                tempFrame = tempFrame.append(tempFrame)
                return tempFrame
        except Exception as exc:
            print("Exception at final output: "+ str(exc))
            return tempFrame
listToAppendEachDict = pd.DataFrame()
listToAppendEachDict = getPageReviews(flipkart_review_page)
listToAppendEachDict.drop_duplicates(inplace=True)
listToAppendEachDict.sort_values(by='productMainRating',ascending=False, inplace=True)
listToAppendEachDict.to_csv("C:\\Users\\pvalluru\\Desktop\\Python_practice\\Pandas\\dataset\\flipkart_iphone_review.csv", index= False)
```
